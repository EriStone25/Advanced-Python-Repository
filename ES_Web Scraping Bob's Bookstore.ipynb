```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://btech-data-analytics.github.io/bridgerland-technical-college/bookstore.html').text
response
```

```python
soup = BeautifulSoup(response, 'html.parser')
```

```python
soup.find('table').find_all('tr', class_='book')[0].find_all('td')[1].text
```

```python
isbn = []
title = []
author = []
language = []
pages = []
topic = []
price = []
for book in soup.find('table').find_all('tr', class_='book'):
    isbn.append(book.find_all('td')[0].text)    
    title.append(book.find_all('td')[1].text)
    author.append(book.find_all('td')[2].text)
    language.append(book.find_all('td')[3].text)
    pages.append(book.find_all('td')[4].text)
    topic.append(book.find_all('td')[5].text)
    price.append(book.find_all('td')[6].text)
```

```python
df = pd.DataFrame({
        'ISBN': isbn,
        'Title': title,
        'Author': author,
        'Language': language,
        'Pages': pages,
        'Topic': topic,
        'Price': price
})
df.head()
```

```python
# Question 1:
# Which author has the most books listed at Bob's Bookstore?
df['Author'].value_counts()
```

```python
# Question 2:
'''Which is the most popular topic among books at Bob's Bookstore
(which topic occurs most frequently)?'''
df['Topic'].value_counts()
```

```python
# Question 3:
# Which topic of books is the most expensive on average?
#df['Price'] = df['Price'].str.replace('$','',regex=False).astype(float).astype(int)
df[['Topic','Price']].groupby('Topic').mean('Price').sort_values(by='Price',ascending=False)
```

```python
# Question 4:
# Which topic of book has the most pages, on average?
df['Pages'] = df['Pages'].astype(int)

df[['Topic','Pages']].groupby('Topic').mean('Pages').sort_values(by='Pages',ascending=False)
```
