```python
import pandas as pd
df = pd.read_csv('companies.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.head()
```

```python
df.info()
```

```python
# Question 1:
'''Using matplotlib/seaborn, create a countplot that shows a count
of how many of the companies in the data set are from each state.
Paste a picture of the graph below. Note that you may need to adjust
the width/height of the graph appropriately.'''
plt.figure(figsize=(12,6))
sns.countplot(x='state',data=df)
plt.show()
```

```python
# Question 2:
# Which state are most of the companies from?
# California
```

```python
# Question 3:
'''Using matplotlib/seaborn, create a bar chart that shows a sum
of how many total job openings there are in each industry.
Paste a picture of the graph below. Note that you may need to adjust
the width/height of the graph appropriately.'''

df_joi = df[['job_openings','Industry']].groupby('Industry').sum('job_openings')
plt.figure(figsize=(20,8))
sns.barplot(x='Industry', y='job_openings', data=df_joi)
plt.ylabel('Total Job Openings')
plt.xlabel('Industry')
plt.show()
```

```python
# Question 4:
# Which industry has the most job openings?
# Fintech
```

```python
# Question 5:
'''Using matplotlib/seaborn, create a line chart that shows the
average growth rate for each founded year.
Paste a picture of the graph below. Note that you may need to adjust
the width/height of the graph appropriately.'''
df_fgp = df[['founded','growth_percentage']].groupby('founded').mean('growth_percentage')

sns.lineplot(x='founded',y='growth_percentage', data=df_fgp,marker='o')
plt.ylabel('Average Growth Rate')
plt.xlabel('Year Founded')
plt.show()
```

```python
# Question 6:
'''From the years 2000 to 2020, does the growth percentage seem
to be increasing, decreasing, or not changing at all?'''
# Increasing
```

```python
# Question 7:
'''Using matplotlib/seaborn, create a scatter plot that compares
the number of employees to estimated revenues for each company.
Paste a picture of the graph below. Note that you may need to adjust
the width/height of the graph appropriately.'''

df_comp = df[['employees','estimated_revenues','company_name']].groupby('company_name').mean('estimated_revenues')
plt.figure(figsize=(20,8))
sns.lmplot(data=df_comp,x='employees',y='estimated_revenues',line_kws={'color':'red'})
plt.xlabel('Number of Employees')
plt.ylabel('Estimated Revenues')
plt.show()
```

```python
# Question 8:
'''What is the relationship between number of employees and
estimated revenues?'''
# As the number of employees increases, so does estimated revenues
```

```python
# Question 9:
'''Using matplotlib/seaborn, create a histogram that measures
the number of companies founded in each year.
Paste a picture of the graph below. Note that you may need to adjust
the width/height of the graph appropriately.'''
df['founded'].nunique()
df_fcn = df[['founded','company_name']].groupby('founded').nunique('company_name')

sns.histplot(x='founded',data=df_fcn,bins=38)
plt.ylabel('Number of Companies')
plt.xlabel('Year Founded')
plt.show()
```

```python
# Question 10:
'''Does the histogram that results from measuring the year each
company was created create a graph that is right skewed, left skewed,
or not skewed at all?'''
# Left skewed
```
